# 1480. Running Sum of 1D Array 

```python
def runningSum(nums):
    running = []
    total = 0
    for n in nums:
        total += n
        running.append(total)
    return running
```

```python
print(runningSum([1, 2, 3, 4]))
```

# 1470. shuffle the code

```python
def shuffle(nums, n):
    result = []
    for i in range(n):
        result.append(nums[i])      # xi
        result.append(nums[i + n])  # yi
    return result
```

```python
print(shuffle([2,5,1,3,4,7], 3))
# [2, 3, 5, 4, 1, 7]

print(shuffle([1,2,3,4,4,3,2,1], 4))
# [1, 4, 2, 3, 3, 2, 4, 1]

print(shuffle([1,1,2,2], 2))
# [1, 2, 1, 2]
```

# 1431. kids with the Greatest Number of Candies

```python
# Kids With the Greatest Number of Candies

def kidsWithCandies(candies, extraCandies):
    max_candy = max(candies)
    return [(c + extraCandies) >= max_candy for c in candies]

# Example 1
candies1 = [2, 3, 5, 1, 3]
extra1 = 3
print("Input:", candies1, "Extra:", extra1)
print("Output:", kidsWithCandies(candies1, extra1))
print()

# Example 2
candies2 = [4, 2, 1, 1, 2]
extra2 = 1
print("Input:", candies2, "Extra:", extra2)
print("Output:", kidsWithCandies(candies2, extra2))
print()

# Example 3
candies3 = [12, 1, 12]
extra3 = 10
print("Input:", candies3, "Extra:", extra3)
print("Output:", kidsWithCandies(candies3, extra3))
```
